## Importation des Donnes

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import train_test_split
```

```python
data=pd.read_csv('train_set.csv')
data=data.drop(['moyeneGeneral','Note1', 'Note2', 'Note3','Note4', 'Note5', 'Note6', 'Note7', 'Note8','Nom','ID'],axis=1)
data.head()
```

## Preprocessing

```python
### creer une colonne pour connaitre l'age
data['age'] = data['DateNaissance'].apply(  
    lambda x: int(x.split("-")[2] if '-' in x else int(x.split(" ")[2]) )
    ).apply(lambda x : 2021 - x if x > 1900  else ( 121 - x if x > 21 else 21 - x))
```

```python
data=data.drop('DateNaissance',axis=1)
```

```python
data=data.drop('Unnamed: 0',axis=1)
```

```python
# Transformer les observations
data['Decision'] = data['Decision'].replace(['Ajourné', 'Sessionnaire', 'Abscent', 'Examen annulé à cause du Téléphone', 'Examen annulé à cause du comportement'], 'Ajourné')
```

```python
### encoder la variable target 
data['Decision']=pd.get_dummies(data['Decision'],drop_first=True)
data.head()
```

```python
data.head()
```

## Preparer les donnes pour le modele

```python
# Sélection des variables catégorielles et numériques
cat_vars = ['Serie,x', 'GroupeAnonymat', 'LieuNaissance', 'Centre', 'Etablissement', 'Willaya', 'moughataa']
num_vars = ['age']
```

```python
# Sélection des données d'entraînement
train_data = data[cat_vars + num_vars]
train_target = data['Decision']
```

```python
# Encodage des variables catégorielles
encoder = OneHotEncoder(sparse=False, handle_unknown='ignore')
X = encoder.fit_transform(train_data[cat_vars])
```

```python
# Concaténation des données catégorielles et numériques
X = np.concatenate([X, train_data[num_vars]], axis=1)
```

## Entrainement du modele

```python
# Séparation des données en ensembles d'entraînement et de test
X_train, X_test, y_train, y_test = train_test_split(X, train_target, test_size=0.2, random_state=42)
```

```python
# Instanciation et entraînement du modèle de régression logistique
clf = LogisticRegression()
clf.fit(X_train, y_train)
```

```python
# Évaluation du modèle sur les données de test
score = clf.score(X_test, y_test)
print(f"Score : {score}")
```

```python
y_pred = clf.predict(X_test)
```

```python
y_pred
```
